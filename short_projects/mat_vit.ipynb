```python
def forward_algorithm(obs_seq, pi, P, E, states, observations):
    T = len(obs_seq)
    K = len(pi)
    forward_prob = np.zeros((K, T))

    # Initialize with 'alternate' state
    forward_prob[:, 0] = np.array([0, 1, 0]) * E[:, observations.index(obs_seq[0])]

    # Recursion
    for t in range(1, T):
        for i in range(K):
            forward_prob[i, t] = sum(forward_prob[j, t-1] * P[j, i] * E[i, observations.index(obs_seq[t])] for j in range(K))

    # Termination
    prob_obs_seq = sum(forward_prob[:, T-1])
    print("Forward Probability Matrix: \n", forward_prob, "\n")
    return prob_obs_seq
```
